```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

# Read the image in grayscale
data = iio.imread('fruit.jpg', pilmode='L')  # Use pilmode='L' for grayscale
print("Image shape:", data.shape)  # Check the shape of the image

# Apply a shift to the image
bdata = nd.shift(data, (50, -30))

# Display the shifted image
plt.imshow(bdata, cmap='gray')
plt.axis('off')  # Turn off axes for better visualization
plt.show()
```

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

data = iio.imread('fruit.jpg')
print("Original shape:", data.shape)

zoom_in_data = nd.zoom(data, (3, 3, 1))
print("Zoomed in shape:", zoom_in_data.shape)

zoom_out_data = nd.zoom(data, (0.3, 0.3, 1))
print("Zoomed out shape:", zoom_out_data.shape)

fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].imshow(zoom_in_data)
axs[0].set_title("Zoomed In (3x)")
axs[0].axis('off')

axs[1].imshow(zoom_out_data)
axs[1].set_title("Zoomed Out (0.3x)")
axs[1].axis('off')

plt.show()
```

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

data = iio.imread('fruit.jpg')
print("Original shape:", data.shape)

rotated_reshape = nd.rotate(data, 45, reshape=True)
print("Rotated shape (reshape=True):", rotated_reshape.shape)

rotated_no_reshape = nd.rotate(data, 45, reshape=False)
print("Rotated shape (reshape=False):", rotated_no_reshape.shape)

fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].imshow(rotated_reshape)
axs[0].set_title("Rotated 45° (reshape=True)")
axs[0].axis('off')

axs[1].imshow(rotated_no_reshape)
axs[1].set_title("Rotated 45° (reshape=False)")
axs[1].axis('off')

plt.show()
```

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

data = iio.imread('world_cup.jpg')

# Dilation với size = (5, 5)
dilate = nd.grey_dilation(data, size=(5, 5, 1))
plt.imshow(dilate)
plt.title("Dilation (5x5)")
plt.show()

# Erosion với size = (5, 5)
erode = nd.grey_erosion(data, size=(5, 5, 1))
plt.imshow(erode)
plt.title("Erosion (5x5)")
plt.show()
```

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

def GeoFun(outcoord):
    a = 5 * np.sin(outcoord[0]/5.0) + outcoord[0]
    b = 5 * np.sin(outcoord[1]/5.0) + outcoord[1]
    return a, b

# Đọc ảnh
data = iio.imread('world_cup.jpg', mode='L')

# Áp dụng biến đổi hình học
d1 = nd.geometric_transform(data, GeoFun)
plt.imshow(d1, cmap='gray')
plt.title("Geometric Transform with sin")
plt.show()
```
